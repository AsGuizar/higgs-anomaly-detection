```python
# ══════════════════════════════════════════════════════════════════
# 01 — Exploratory Data Analysis
# Higgs Boson Anomaly Detection Project
#
# Research context:
# The HIGGS dataset (Baldi et al., 2014) contains 11 million simulated
# particle collision events. Each event is described by 28 kinematic
# features — 21 low-level detector measurements and 7 high-level features
# derived by physicists from theory.
#
# In this notebook we:
# 1. Load and inspect the dataset
# 2. Understand class imbalance
# 3. Visualize feature distributions and separability
# 4. Identify which features contain the most discriminating information
#
# Key framing: Labels are available in this dataset but will be
# withheld from all models during training. We use them here only to
# understand the data — and during final evaluation to measure performance.
# ══════════════════════════════════════════════════════════════════


# ── 0. SETUP ──────────────────────────────────────────────────────

!pip install ucimlrepo seaborn -q
!git clone https://github.com/AsGuizar/higgs-anomaly-detection.git
%cd higgs-anomaly-detection

import os, sys
os.makedirs('results/figures', exist_ok=True)
sys.path.insert(0, '.')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sklearn.metrics import roc_auc_score

from src.preprocess import LOW_LEVEL_FEATURES, HIGH_LEVEL_FEATURES, ALL_FEATURES
from src.visualize import (
    plot_class_balance, plot_feature_distributions,
    plot_feature_separability, plot_correlation_matrix
)

plt.style.use('seaborn-v0_8-whitegrid')
print("Imports OK")


# ── 1. LOAD DATASET ───────────────────────────────────────────────
# Direct download from UCI file server — no API, no library.
# First run downloads ~2.6GB and caches it in data/.
# Every run after that loads from the cached file.

DATA_PATH = Path('data/HIGGS.csv.gz')

if not DATA_PATH.exists():
    print("Downloading HIGGS dataset (~2.6GB, takes a few minutes)...")
    !wget -O data/HIGGS.csv.gz https://archive.ics.uci.edu/ml/machine-learning-databases/00280/HIGGS.csv.gz
    print("Download complete.")

print("Loading dataset...")
col_names = ['label'] + ALL_FEATURES
df = pd.read_csv(DATA_PATH, header=None, names=col_names)
df['label'] = df['label'].astype(int)

print(f"Shape:             {df.shape}")
print(f"Features:          {len(ALL_FEATURES)}")
print(f"  Low-level:       {len(LOW_LEVEL_FEATURES)}")
print(f"  High-level:      {len(HIGH_LEVEL_FEATURES)}")
print(f"\nSignal events:     {(df.label==1).sum():>10,}  ({(df.label==1).mean()*100:.2f}%)")
print(f"Background events: {(df.label==0).sum():>10,}  ({(df.label==0).mean()*100:.2f}%)")


# ── 2. CLASS BALANCE ──────────────────────────────────────────────
# ~47% signal, ~53% background.
# This is simulated data — real detector data would be far more extreme.
# In production searches, signal events may be 1 in 10^9 or rarer.
# Our balanced evaluation set assesses model behavior at a meaningful fraction.

plot_class_balance(
    df['label'].values,
    save_path='results/figures/01_class_balance.png'
)


# ── 3. BASIC STATISTICS ───────────────────────────────────────────
# Compare summary statistics of signal vs background across all features.

print("── Background events ──")
display(df[df.label==0][ALL_FEATURES].describe().round(3))

print("── Signal events ──")
display(df[df.label==1][ALL_FEATURES].describe().round(3))


# ── 4. FEATURE DISTRIBUTIONS ─────────────────────────────────────
# Sample 200k events for visualization speed.
# High-level features (yellow) should show cleaner signal/background
# separation — physicists engineered them specifically for that.
#
# Key question for our anomaly detectors:
# can they rediscover this structure without being told which features matter?

viz_sample = df.sample(n=200_000, random_state=42)

plot_feature_distributions(
    viz_sample,
    label_col='label',
    feature_cols=ALL_FEATURES,
    low_level_n=21,
    save_path='results/figures/01_feature_distributions.png'
)


# ── 5. FEATURE SEPARABILITY ───────────────────────────────────────
# For each feature, compute single-feature AUC against the label.
# This tells us which features carry the most discriminating information.
#
# Finding: high-level features cluster at the top of the ranking.
# Theory-derived features encode more discriminating information
# than raw detector measurements.
#
# We return to this in Notebook 04 — does the autoencoder's reconstruction
# error profile match this ranking without being told to care about it?

sep_series = plot_feature_separability(
    viz_sample,
    label_col='label',
    feature_cols=ALL_FEATURES,
    high_level_features=HIGH_LEVEL_FEATURES,
    save_path='results/figures/01_feature_separability.png'
)

print("\nTop 7 most separating features:")
print(sep_series.tail(7).round(4).to_string())
print("\nBottom 7 least separating features:")
print(sep_series.head(7).round(4).to_string())


# ── 6. CORRELATION STRUCTURE ──────────────────────────────────────
# Signal and background have different covariance structures —
# especially in the high-level feature block (bottom right quadrant).
#
# This matters for the autoencoder: signal events are not just wrong
# in individual feature values — their relationships are wrong too.
# An autoencoder trained on background should fail on signal for both reasons.

bg_sample  = df[df.label==0].sample(n=50_000, random_state=42)
sig_sample = df[df.label==1].sample(n=50_000, random_state=42)

plot_correlation_matrix(
    bg_sample,
    feature_cols=ALL_FEATURES,
    low_level_n=21,
    title_suffix='— Background Events',
    save_path='results/figures/01_correlation_matrix_background.png'
)

plot_correlation_matrix(
    sig_sample,
    feature_cols=ALL_FEATURES,
    low_level_n=21,
    title_suffix='— Signal Events',
    save_path='results/figures/01_correlation_matrix_signal.png'
)


# ── 7. SAVE PROCESSED SPLITS ──────────────────────────────────────
# Runs download.py which saves:
#   data/background_sample.csv  — 500k background events (training pool)
#   data/test_set.csv           — 100k balanced test set
#   data/test_labels.npy        — labels for evaluation only
# Run once. All subsequent notebooks load from these files.

!python data/download.py


# ══════════════════════════════════════════════════════════════════
# EDA SUMMARY
#
# Finding                              Implication for modeling
# ─────────────────────────────────────────────────────────────────
# High-level features more separating  AE may concentrate error there
# Signal has different covariance      AE trained on BG should fail on signal
# ~47% signal in full dataset          Use balanced test set, report PR-AUC
# No missing values                    No imputation needed
#
# Next: 02_isolation_forest.py — first anomaly detector
# ══════════════════════════════════════════════════════════════════
```
